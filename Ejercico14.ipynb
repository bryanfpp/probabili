## Se va a formar un portafolio de inversión de cuatro instrumentos financieros a partir de cinco de renta fija y seis de renta variable. ¿De cuántas maneras podría formarse el portafolio si:

- a) No hay restricciones?
- b) Debe haber dos instrumentos de renta fija y dos de renta variable?
- c) No debe haber ningún instrumento de renta fija?
- d) Debe haber tres instrumentos de renta fija?
- e) Calcula las probabilidades para los incisos anteriores.
- f) Investiga qué es un instrumento de renta fija y otro de renta variable.
- g)Reflexiona sobre la importancia del ahorro y comparte con tus compañeros de clase las conclusiones a las que llegaste.

## inciso (a

## a) No hay restricciones:
En este caso, podemos elegir cualquier combinación de cuatro instrumentos de la lista total de 11 (5 RF + 6 RV). Utilizamos la fórmula de combinaciones para calcularlo:

$C(11, 4) = \frac{{11!}}{{4! \cdot (11 - 4)!}} = \binom{{11}}{{4}} = 330$


## b) Debe haber dos instrumentos de renta fija y dos de renta variable:
Para cumplir con esta restricción, seleccionamos 2 instrumentos de renta fija de los 5 disponibles (C(5, 2) = 10) y 2 instrumentos de renta variable de los 6 disponibles (C(6, 2) = 15). Luego multiplicamos estos dos resultados:

$C(5, 2) \cdot C(6, 2) = \frac{{5!}}{{2! \cdot (5 - 2)!}} \cdot \frac{{6!}}{{2! \cdot (6 - 2)!}} = \binom{{5}}{{2}} \cdot \binom{{6}}{{2}} = 10 \cdot 15 = 150$

## c) No debe haber ningún instrumento de renta fija:
En este caso, solo podemos elegir 4 instrumentos de renta variable de los 6 disponibles (C(6, 4) = 15).


$C(6, 4) = \binom{{6}}{{4}} = 15$

## d) Debe haber tres instrumentos de renta fija:
Para cumplir con esta restricción, seleccionamos 3 instrumentos de renta fija de los 5 disponibles (C(5, 3) = 10) y 1 instrumento de renta variable de los 6 disponibles (C(6, 1) = 6). Luego multiplicamos estos dos resultados:


$C(5, 3) \cdot C(6, 1) = \frac{{5!}}{{3! \cdot (5 - 3)!}} \cdot \frac{{6!}}{{1! \cdot (6 - 1)!}} = \binom{{5}}{{3}} \cdot \binom{{6}}{{1}} = 10 \cdot 6 = 60$

## e) Calcula las probabilidades para los incisos anteriores.

1. a) Probabilidad sin restricciones:
$P(a) = \frac{330}{330} = 1$

2. b) Probabilidad con dos instrumentos de renta fija y dos de renta variable:
$P(b) = \frac{150}{330} \approx 0.4545$

3. c) Probabilidad sin ningún instrumento de renta fija:
$P(c) = \frac{15}{330} \approx 0.0455$

4. d) Probabilidad con tres instrumentos de renta fija:
$P(d) = \frac{60}{330} \approx 0.1818$


# f) Investiga qué es un instrumento de renta fija y otro de renta variable.

### Un instrumento de renta variable:
un tipo de inversión en la que se presta dinero a un emisor (como un gobierno o una empresa) a cambio de pagos regulares de intereses y la devolución del capital prestado en una fecha futura. Los instrumentos de renta fija incluyen bonos, pagarés, letras del tesoro y otros valores de deuda.

### Un instrumento de renta variable:

representa una participación en la propiedad de una empresa. Los inversionistas que poseen acciones de una empresa son propietarios parciales de la misma y tienen derecho a recibir parte de las ganancias y a participar en la toma de decisiones de la empresa a través de su derecho a voto en las juntas de accionistas.
